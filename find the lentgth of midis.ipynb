```python
from music21 import converter
import os

def get_total_bars(midi_file_path):
    # Parse the MIDI file
    midi_stream = converter.parse(midi_file_path)

    # Extract the time signature
    time_signature = None
    for element in midi_stream.flat:
        if 'TimeSignature' in element.classes:
            time_signature = element
            break

    if time_signature is None:
        raise ValueError(f"No time signature found in {midi_file_path}.")

    # Calculate the total duration in quarter notes
    total_duration = midi_stream.highestTime

    # Calculate the total length in bars
    beats_per_measure = time_signature.numerator
    quarter_notes_per_beat = 4  # Assuming quarter note as the basic unit
    total_bars = total_duration / (beats_per_measure * quarter_notes_per_beat)

    return total_bars

# Example usage:
midi_folder_path = r'C:\Models\data_pre_processing_hihat\hi_hat_midis_monophonic'
for root, dirs, files in os.walk(midi_folder_path):
    k = 0
    for file in files:
        if file.endswith(".mid"):
            midi_file_path = os.path.join(root, file)
            total_bars = get_total_bars(midi_file_path)

            # Check if total bars is not equal to 1 or 2
            if total_bars not in [1, 2]:
                print(f"Deleting {midi_file_path} (Total bars: {total_bars})")
                os.remove(midi_file_path)
                k += 1
            else:
                print(f"{midi_file_path}: Total bars - {total_bars}")
    print(f"total number of deleted files:{k}")
```
